```python
df['arrival_date']=pd.to_datetime(df[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']].astype(str).agg('-'.join, axis=1), format='%Y-%B-%d')
```

```python
df_columns=df.columns
df_columns
```

```python
nan_list=[]
for column_name in df.columns:
    if df[column_name].isnull().any():
        print(f"{column_name} True")
        nan_list.append(f"{column_name}")
    else:
        print(f"{column_name} False")
```

```python
nan_list
```

```python
for i in nan_list:
    df[i]=df[i].replace("nan",None)
```

```python
df
```

```python
drop_column=['arrival_date_week_number','stays_in_weekend_nights','stays_in_week_nights','arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month','credit_card','phone-number','email','assigned_room_type','reservation_status_date','previous_bookings_not_canceled','agent','company','booking_changes','days_in_waiting_list','distribution_channel']
for i in drop_column:
    df=df.drop(columns=[i])
```
